```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel("company_sales_data.xlsx")

# Exercise 1: Line plot of Total Profit
plt.plot(data['month_number'], data['total_profit'], marker='o', color='b')
plt.title("Company Profit per Month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()
```

```python
# Exercise 2: Subplots for Bathing Soap and Facewash
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Bathing soap
axs[0].plot(data['month_number'], data['bathingsoap'], marker='o', color='g')
axs[0].set_title("Bathing Soap Sales Data")
axs[0].set_ylabel("Units Sold")
axs[0].grid(True)

# Facewash
axs[1].plot(data['month_number'], data['facewash'], marker='o', color='r')
axs[1].set_title("Facewash Sales Data")
axs[1].set_xlabel("Month Number")
axs[1].set_ylabel("Units Sold")
axs[1].grid(True)

plt.tight_layout()
plt.show()

```
